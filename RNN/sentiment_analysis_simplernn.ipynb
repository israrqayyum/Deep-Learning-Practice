```python
docs = ['Hi Everyone',
		'This is Python Notebook',
		'hip hip hurray',
		'OpenAI lanuches a new browser Atlas',
		'Everyday there is new advancement in cutting edge technologies',
		'World is changing']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(docs)
```

```python
len(tokenizer.word_index)
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences
sequences = pad_sequences(sequences,padding='post')
sequences
```

```python
# model = Sequential()
# model.add(Embedding(17,output_dim=2,input_length=5))

# model.summary()
from keras import Sequential
from keras.layers import Embedding

model = Sequential()
model.add(Embedding(input_dim=23, output_dim=2, input_shape=(9,)))

model.summary()
```

```python
model.compile('adam','accuracy')
```

```python
pred = model.predict(sequences)
print(pred)
```

```python
from keras.datasets import imdb
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

```python
X_train = pad_sequences(X_train,padding='post',maxlen=50)
X_test = pad_sequences(X_test,padding='post',maxlen=50)
```

```python
X_train.shape
```

```python
# model = Sequential()
# model.add(Embedding(10000, 2,50))
# model.add(SimpleRNN(32,return_sequences=False))
# model.add(Dense(1, activation='sigmoid'))

# model.summary()

model = Sequential()
model.add(Embedding(input_dim=10000, output_dim=2, input_shape=(50,)))  # 50 = sequence length
model.add(SimpleRNN(32, return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(X_train, y_train,epochs=5,validation_data=(X_test,y_test))
```
